### Task 2.1

Read in Bloom_etal_2018_Reduced_Dataset.csv and create two new columns ('genus','species') that consists of the column taxa split at the underscore. Print out the head of this new dataframe and the number of unique genera**

hint: pd.str.split(,expand=True)

for example:

taxa	genus	species
Alosa_alabamae	Alosa	alabamae


```python
bloom_df = pd.read_csv('Bloom_etal_2018_Reduced_Dataset.csv')
```

```python
bloom_df.head()
```

```python
bloom_df.at[49, 'taxa'] = 'Thryssa_dussumieri'
```

above code fixes an error in a single cell that is index number 49 in taxa column 

```python
bloom_df[['Genus', 'Species']] = bloom_df['taxa'].str.split('_', expand=True)
```

```python
bloom_df.head()
```

```python
bloom_df['Genus'].value_counts()
```

```python
cnt = 0
visited = []
for i in range(0, len(bloom_df['Genus'])):
        if bloom_df['Genus'][i] not in visited:
                visited.append(bloom_df['Genus'][i])
                cnt += 1
print("Number of unique genera :",
      cnt)
print("unique values :",
     visited)
```
